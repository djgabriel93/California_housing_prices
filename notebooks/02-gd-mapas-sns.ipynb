```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from src.config import DADOS_LIMPOS
from src.graficos import PALETTE, SCATTER_ALPHA

sns.set_theme(palette="bright", style='dark')
```

```python
df = pd.read_parquet(DADOS_LIMPOS)

df.head()
```

```python
#mapa de localizao dos imoveis
fig,ax = plt.subplots(figsize=(10,6))

sns.scatterplot(data=df, x= 'longitude',y = 'latitude', ax=ax, alpha=SCATTER_ALPHA)

plt.show();
```

```python
#mapa de localizao dos imoveis com histograma
sns.jointplot(data=df, x= 'longitude',y = 'latitude', alpha=SCATTER_ALPHA);
```

```python
#mapa de localizao dos imoveis com a proximidade para o Oeceano
fig,ax = plt.subplots(figsize=(10,6))

sns.scatterplot(data=df, x= 'longitude',y = 'latitude', ax=ax, alpha=SCATTER_ALPHA, hue='ocean_proximity')

plt.show();
```

```python
#mapa de localizao dos imoveis com as diferentes rendas

fig,ax = plt.subplots(figsize=(10,6))

sns.scatterplot(data=df, x= 'longitude',y = 'latitude', ax=ax, alpha=0.5, hue='median_income_cat', palette= PALETTE)

plt.show();
```

o Mapa acima representa a renda por domicilios no estado da California.
- As maiores rendas se concentram prximos as regies de Los Angeles e So Francisco e proximidades.
- As rendas mdias se concentram entre essas regies de alta renda e prximas ao oceano.
- As rendas menores esto concentrada mais para o interior do estado.

```python
#mapa de localizao dos imoveis com a escala de preos das casas em outras cores
fig, ax = plt.subplots(figsize=(10, 6))

scatter = sns.scatterplot(
    data=df, 
    x='longitude', 
    y='latitude', 
    hue='median_house_value', 
    palette=PALETTE,     
    alpha=0.5,          
    ax=ax
)

norm_median_house_value = plt.Normalize(df['median_house_value'].min(), df['median_house_value'].max())
sm_median_house_value = plt.cm.ScalarMappable(norm_median_house_value, cmap = PALETTE)
fig.colorbar(sm_median_house_value, ax=ax)
ax.get_legend().remove()

# Ttulos e labels
ax.set_title('Distribuio de Preos de Imveis por Localizao', fontsize=15, pad=20)

plt.tight_layout()
plt.show()
```

No mapa com os preos das casas podemos observar que as reas mais caras esto localizadas prximos as mesmas regies que tem a renda mais alta.
Isso faz sentido pois vimos na analise exploratria que existe uma forte correlao entre essas 2 variveis.
